### Requests 

```python
import requests as re

req = re.get('https://www.forbes.pl/diamenty/najszybciej-rozwijajace-sie-firmy-w-wojewodztwie-mazowieckim-w-2020-roku/vbbfems')

if req.status_code == 200:
    print('Strona pobrana')
else:
    print('Błąd')
```

```python
req.headers
```

```python
#modyfikacja user-agent
req.request.headers
```

```python
print(req.text)
```

```python
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
div = soup.find('div',class_='tableWrapper')
divs = div.find_all('td')


forbes_dict = {'Firma':[],'Siedziba':[],'Zysk':[],'Średnia':[],'Branża':[],'Sprzedaż 2019':[]}
for div in divs:
    
    if str(div).find("Siedziba") != -1:
        forbes_dict['Siedziba'].append(div.text.strip())
    if str(div).find("Firma") != -1:
        forbes_dict['Firma'].append(div.text.strip())
    if str(div).find("Zysk") != -1:
        forbes_dict['Zysk'].append(div.text.strip())
    if str(div).find("Branża") != -1:
        forbes_dict['Branża'].append(div.text.strip())
    if str(div).find("Średnia") != -1:
        forbes_dict['Średnia'].append(div.text.strip())
    if str(div).find("Sprzedaż 2019") != -1:
        forbes_dict['Sprzedaż 2019'].append(div.text.strip())



data=pd.DataFrame(forbes_dict)
```

### json

```python
x = [{i: j for i in range(5)} for j in range(10)]
x
```

```python
import json
json.dumps(x)
```

```python
with open('test.json','w') as file:
    json.dump(x, file)
```

```python
with open('test.json','r') as file:
    y = json.load(file)
    print(y)
```

```python
import gzip

with gzip.open('test.json.gzip','wt', encoding = 'utf-8') as file:
    json.dump(x, file)
```

```python
with gzip.open('test.json.gzip','r') as file:
    y = json.load(file)
    print(y)
```
